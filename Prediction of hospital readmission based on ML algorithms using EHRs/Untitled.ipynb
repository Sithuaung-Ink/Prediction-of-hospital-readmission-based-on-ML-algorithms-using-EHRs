```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

x = np.arange(0, 20)
m = 1.5
c = 2
y = m*x + c
y_2 = m*x + 2*c

font_label = {'family': 'Times New Roman',
        'color':  'black',
        'weight': 'normal',
        'size': 12}

font_legend = font_manager.FontProperties(family='Times New Roman',
                                   weight='normal',
                                   style='normal', size=12)

plt.figure(dpi=100)
plt.plot(x, y, 'g-D',label='y_1')
plt.plot(x,y_2,'b-o' ,label ='y_2')
plt.title("y = mx+c", fontdict=font_label)
plt.xlabel('x-axis', fontdict=font_label)
plt.xticks(family='Times New Roman', size=12)
plt.ylabel('y-axis', fontdict=font_label)
plt.yticks(family='Times New Roman', size=12)
plt.legend(prop = font_legend, bbox_to_anchor =(0.5,-0.37), loc='lower center')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

x = np.arange(0, 10)
x_1 = df['SVM_AUC']
x_2 = df['KNN_AUC']
x_3 = df['RF_AUC']
x_4 = df['LogReg_AUC']
x_5 = df['XGBoost_AUC']
x_6 = df['LGBM_AUC']
x_7 = df['CatBoost_AUC']

font_label = {'family': 'Times New Roman',
        'color':  'black',
        'weight': 'normal',
        'size': 12}

font_legend = font_manager.FontProperties(family='Times New Roman',
                                   weight='normal',
                                   style='normal', size=12)
#plt.figure(dpi=120)
plt.plot(x_1, 'b-o',label='SVM')
plt.plot(x_2,'g-o' ,label ='KNN')
plt.plot(x_3, 'r-o',label='RF')
plt.plot(x_4,'c-o' ,label ='LogReg')
plt.plot(x_5, 'm-o',label='XGBoost')
plt.plot(x_6,'y-o' ,label ='LGBM')
plt.plot(x_7, 'b-o',label='CatBoost')
plt.grid(axis = 'y')
#plt.title("y = mx+c", fontdict=font_label)
plt.xlabel('Number of features', fontdict=font_label)
plt.xticks(np.arange(len(x)), np. arange(1, len(x)+1), family='Times New Roman', size=12)
plt.ylabel('AUC', fontdict=font_label)
plt.yticks(family='Times New Roman', size=12)
#plt.legend(prop = font_legend, bbox_to_anchor =(1,0.5), loc='center left')
plt.legend(prop = font_legend, loc='lower right')
plt.savefig("AUC.png",dpi=300)
plt.show()
```
